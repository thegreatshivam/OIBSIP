## Importing the dataset

```python
import pandas as pd
df = pd.read_csv('CarPrice.csv')
```

```python
df.info()
```

## Converting text to number (four -> 4)

```python
print(df['cylindernumber'].unique())
df['cylindernumber'] = df['cylindernumber'].replace({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8})
```

```python
print(df['doornumber'].unique())
df['doornumber'] = df['doornumber'].replace({'four':4, 'two':2})
```

## Remove the insignificant column from the dataset

```python
df = df.drop(['car_ID','CarName'], axis=1)
df.head()
```

```python
categories = df.select_dtypes('object').columns
categories
```

## Splitting the features and target columns

```python
y = df['price']
y
```

```python
X = df.drop('price', axis=1)
X
```

```python
X.head()
```

## Encoding all categorical data

```python
X = pd.get_dummies(X)
X
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
X.head()
```

## Splitting the dataset into test and split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

## Creating and training the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

## Predicting the output and checking the accuracy of model

```python
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score

accuracy = r2_score(y_test, pred) * 100
print("The model is {0:.2f}% accurate.".format(accuracy))
```
